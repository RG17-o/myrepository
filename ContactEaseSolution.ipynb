```python
import sys
import json
import os



#Nome del file dei contatti
FILE_CONTACTS= "contacts.json"

#Dizionario dei contatti
Contacts={}

def load_contacts(file_name=FILE_CONTACTS):

    """Carica i contatti dal file che viene indicato."""
    global Contacts
    if os.path.exists(file_name):
        with open(file_name,"r") as f:
            Contacts= json.load(f)
    else:
        Contacts = {}

def save_contacts(file_name=FILE_CONTACTS):
    """Funzione per salvare i contatti nel file specificato."""
    with open(file_name,"w") as f:
        json.dump(Contacts , f)

def add_contact():
    """Aggiunge un contatto nuovo."""
    try:
        if len(sys.argv) < 5:
            print("Usa il comando nel formato: AGGIUNGI nome cognome numero")
            return

        nome_e_cognome = sys.argv[2].upper() + " " + sys.argv[3].upper()
        numero = sys.argv[4]

        if not (numero.isdigit() and len(numero) == 10):
            print("Il numero deve essere formato da 10 cifre")
        else:
            global Contacts
            Contacts[nome_e_cognome] = numero
            print(f"Contatto aggiunto: {nome_e_cognome} -> {numero}")

        return Contacts
    except Exception as e:
        print(f"Errore: {e}")
        print("Scrivere esattamente nel formato: AGGIUNGI nome cognome numero")

def change_contact():
    """Modifica un contatto che già esiste."""
    try:
        if sys.argv[2].upper() =="RINOMINA":
            vecchio_nome= sys.argv[3].upper() + " " + sys.argv[4].upper()
            nuovo_nome= sys.argv[5].upper() + " " + sys.argv[6].upper()
            if vecchio_nome not in Contacts :
                print("Il vecchio nome non esiste")
            else:
                valore_da_preservare = Contacts[vecchio_nome]
                Contacts[nuovo_nome]=valore_da_preservare
                del Contacts[vecchio_nome]

        elif sys.argv[2].upper()=="CAMBIA":
            nuovo_numero= int(sys.argv[8])
            contatto_da_cambiare= sys.argv[5].upper()+" "+ sys.argv[6].upper()
            Contacts[contatto_da_cambiare]=nuovo_numero
    except Exception as e:
        print(f"Errore: {e}")
        print("Operazione non andata a buon fine: scrivere esattamente nel formato: 'MODIFICA RINOMINA VECCHIO_NOME VECCHIO_COGNOME NUOVO_NOME NUOVO_COGNOME' o 'MODIFICA CAMBIA NUMERO DI NOME_COGNOME IN NUOVO_NUMERO' ")

def delete_contact():
    """Elimina un contatto esistente."""
    try:
        nome_cognome= sys.argv[2].upper()+ " " + sys.argv[3].upper()
        if nome_cognome not in Contacts :
            print("Il contatto inserito non esiste")
        else:
            del Contacts[nome_cognome]
            print("contatto eliminato con successo")
    except Exception as e:
        print(f"Errore : {e}")
        print("Scrivere esattamente nel formato: ELIMINA nome cognome")

def search_contact():
    """Ricerca un contatto."""
    try:
        nome_cognome_search = sys.argv[2].upper()+ " " + sys.argv[3].upper()
        if nome_cognome_search not in Contacts :
            print("Il contatto inserito non esiste")
        else:
            print(f" Il numero di {nome_cognome_search} è {Contacts[nome_cognome_search]}")
    except Exception as e:
        print(f"Errore : {e}")
        print("Scrivere esattamente nel formato: RICERCA nome cognome ")


def main():
    """Funzione principale del programma creato."""
    file_name= FILE_CONTACTS
    if len(sys.argv) > 2 and sys.argv[1].upper() == "FILE":
      file_name = sys.argv[2]
      del sys.argv[1:3]

    load_contacts(file_name)

    while True:
            print("Cosa desideri fare?\nAGGIUNGI : nome cognome numero\nMOSTRA\nMODIFICA : 'MODIFICA RINOMINA VECCHIO_NOME VECCHIO_COGNOME NUOVO_NOME NUOVO_COGNOME' / 'MODIFICA CAMBIA NUMERO DI NOME_COGNOME IN NUOVO_NUMERO'\nELIMINA nome cognome\nRICERCA nome cognome\nESCI")
            command = input().upper().split()

            if not command:
                print("Inserisci un comando valido.")
                continue

            sys.argv = ["main.py"] + command

            if command[0] == "AGGIUNGI":
                add_contact()
            elif command[0] == "MOSTRA":
                print(Contacts)
            elif command[0] == "MODIFICA":
                change_contact()
            elif command[0] == "ELIMINA":
                delete_contact()
            elif command[0] == "RICERCA":
                search_contact()
            elif command[0] == "ESCI":
                print("Uscita dal programma.")
                break
            else:
                print("Comando non riconosciuto. Riprova.")

    save_contacts(file_name)

if __name__ == "__main__":
    main()
```
